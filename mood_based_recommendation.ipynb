```python
import pandas as pd
import random
import music
import numpy as np
from numpy.linalg import norm
```

```python
#read the dataset generated
data=pd.read_csv('songs_mood.csv')
data.head()
```

```python
#create a column for essential features to identify the moods
data['mood_feature']=data[['valence','energy']].values.tolist()
data.head()
```

```python
def track_id(song_name):
    p=str(data[data.track_name==song_name]['id'])
    c=p[8:30]
    return c
    
```

```python
log_in = music.authenticate()
```

```python
#function for recommend music based on mood
def recommendation(song_name,login_id=log_in,num=5):
    track_id='5LkAwO0AkY6Z3M4bCadu4r'
    #print(track_id)
    audio_features=login_id.track_audio_features(track_id)
    mood_features = np.array([audio_features.valence, audio_features.energy])
    data["distance"] = data["mood_feature"].apply(lambda x: norm(mood_features-np.array(x)))
    sort_data = data.sort_values(by = "distance", ascending = True,inplace=False,axis=0,ignore_index=False)
    sort_data = sort_data[sort_data["id"] != track_id]
    return sort_data.iloc[:num]
    
```

```python
#test the function
song_name='Quicksand'
recommendation(song_name,log_in)
```
